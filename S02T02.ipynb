```python
###########################################################################################
###############  Aquí comença el codi del programa ########################################
###########################################################################################

img_array_red=np.copy(img_array) # Fem una copia de la imagen per preservar l'original
img_array_red[:,:,1]=0
img_array_red[:,:,2]=0
plt.suptitle("\nThe Mask")
plt.subplot(1, 4, 1)
plt.title("Canal Vermell")
plt.imshow(img_array_red)
plt.axis('off')
print("\nLa imatge VERMELLA s'ha desat a la carpeta del NoteBook amb el nom: \"TheMask_RED_savefig.png\"")
plt.savefig("TheMask_RED_savefig.png", dpi=300)
plt.close()

img_array_green=np.copy(img_array) # Fem una copia de la imagen per preservar l'original
img_array_green[:,:,0]=0
img_array_green[:,:,2]=0
plt.suptitle("\nThe Mask")
plt.subplot(142)
plt.title("Canal Verd")
plt.imshow(img_array_green)
plt.axis('off')
print("\nLa imatge VERDA s'ha desat a la carpeta del NoteBook amb el nom: \"TheMask_GREEN_savefig.png\"")
plt.savefig("TheMask_GREEN_savefig.png", dpi=300)
plt.close()

img_array_blue=np.copy(img_array) # Fem una copia de la imagen per preservar l'original
img_array_blue[:,:,0]=0
img_array_blue[:,:,1]=0
plt.suptitle("\nThe Mask")
plt.subplot(143)
plt.title("Canal Blau")
plt.imshow(img_array_blue)
plt.axis('off')
print("\nLa imatge BLAVA s'ha desat a la carpeta del NoteBook amb el nom: \"TheMask_BLUE_savefig.png\"")
plt.savefig("TheMask_BLUE_savefig.png", dpi=300)
plt.close()

img_array_mix=np.copy(img_array) # Fem una copia de la imagen per preservar l'original
img_array_mix[127,127,0]=0.5
img_array_mix[255,255,0]=0.5
img_array_mix[:,:,2]=1
plt.suptitle("\nThe Mask")
plt.subplot(1, 4, 4)
plt.title("Canal Mix")
plt.imshow(img_array_mix);       # Posant un ";" al final, eliminem el text de info de la imatge... 
plt.axis('off');
print("\nLa imatge MIX s'ha desat a la carpeta del NoteBook amb el nom: \"TheMask_MIX_savefig.png\"")
plt.savefig("TheMask_MIX_savefig.png", dpi=300)
plt.close()

plt.suptitle("\nThe Mask")
plt.subplot(1, 4, 1)
plt.title("Canal Vermell")
plt.imshow(img_array_red)
plt.axis('off')      # ... però no he aconseguit ajustar els eixos...   plt.axis('off') --> és un "parche"
plt.subplot(1, 4, 2)
plt.title("Canal Verd")
plt.imshow(img_array_green)
plt.axis('off')
plt.subplot(1, 4, 3)
plt.title("Canal Blau")
plt.imshow(img_array_blue)
plt.axis('off')
plt.subplot(1, 4, 4)
plt.title("Canal Mix")
plt.imshow(img_array_mix)
plt.axis('off');

###########################################################################################
###############  Fins aquí el codi del programa ###########################################
###########################################################################################

##############  Bussejant per la Web vaig trobar aquest fragment de codi ##################
###################  ... i em va semblar interessant explorar-lo...   #####################


from matplotlib import pyplot as plt             
import random
import numpy as np

def toggle_cell_code(button_id):
    """
    Adds a button to toggle (show/hide) the code cell but not the output. 
    
    Parameters
    ----------
    button_id : str
        An identifier for cells that will hide/show when button is pressed.
    """
    
    from IPython.display import display_html
    my_html = '''
    <button type="button" id="%s" onclick="code_toggle('%s')">Mostra/Oculta Codi</button>
    <script>
    function code_toggle(my_id) {
        // get the parent element for the cell code and output
        var p = $("#"+my_id);
        if (p.length==0) return;
        while (!p.hasClass("cell")) {
            p = p.parent();
            if (p.prop("tagName") =="body") return;
        }
    // get the cell code and toggle its value
    var cell_code = p.find(".input");
    cell_code.toggle();
    }
    </script>
    ''' %(button_id, button_id)
    return display_html(my_html, raw=True)
    
toggle_cell_code("un_string_unico_y_reconocible")
```
